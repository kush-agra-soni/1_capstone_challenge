##Creating a shipping cost calculator!

You need to create a program that calculates the cost of shipping a package based on its weight and destination

Domestic shipping within the country :

* Packages weighing up to 1 kg cost Rs.50 to ship
Each additional kg costs Rs.10 extra

International shipping to other countries:

* Packages weighing up to 1 kg cost Rs.500 to ship

Each additional kg costs Rs.100 extra
Ready to ship?

```python
weight = float(input("Enter the weight of the package in kg: "))
destination = input("Enter the destination ('domestic' or 'international'): ")

if destination.lower() == 'domestic':
    if weight <= 1:
        shipping_cost = 50
    else:
        shipping_cost = 50 + 10 * (weight - 1)
elif destination.lower() == 'international':
    if weight <= 1:
        shipping_cost = 500
    else:
        shipping_cost = 500 + 100 * (weight - 1)
else:
    print("Invalid destination. Please enter 'domestic' or 'international'.")
    shipping_cost = None

if shipping_cost is not None:
    print(f"The shipping cost for a {weight} kg package to {destination} destination is Rs. {shipping_cost}.")
```

```python
# List of numbers
numbers = [
    [2, 5, 11, 20, 8],
    [9, 4, 15, 28, 17],
    [1, 6, 21, 18, 3],
    [10, 13, 25, 33, 30],
    [14, 7, 16, 19, 22]
]

even_lst = [ ]
odd_lst = [ ]

for lst in numbers:
  for items in lst:
    if items % 2 == 0:
      even_lst.append(items)
    else:
      odd_lst.append(items)

print("Even number = ",len(even_lst))
print("Odd number = ",len(odd_lst))
```

The sequence starts with 0 and 1, and the next number in the sequence is the sum of the previous two numbers.

```python
num = int(input("Enter the number of terms: "))

# Step 2: Initialize the first two terms
a = 0
b = 1

fib_series = []

# Adding the first two terms to the list
if num >= 1:
    fib_series.append(a)
if num >= 2:
    fib_series.append(b)

count = 2  # Since we already have the first two terms
while count < num:
    c = a + b
    fib_series.append(c)

    # Update terms
    a = b
    b = c
    count += 1

print("Fibonacci sequence:")
print(fib_series)
```

## **Counting the Digit**

You have been given a task to write a Python program that counts the number of times a specific digit appears in a range of numbers. You need to use a for loop to iterate over the range of numbers, and a while loop to check each digit in the number.

Example:

* Enter start number: 100
* Enter end number: 150
* Enter digit to count: 5
* The digit 5 appears 6 times between 100 and 150.


```python
# Input the range (start and end) and the digit to count
start_num = int(input("Enter starting number: "))
end_num = int(input("Enter ending number: "))
digit_to_count = input("Enter digit to count: ")

count = 0

# Use a for loop to iterate over the range of numbers
for number in range(start_num, end_num + 1):

    # Use a while loop to check each digit in the current number
    while number > 0:
        digit = number % 10
        if str(digit) == digit_to_count:
            count += 1
        number //= 10


print(f'The digit {digit_to_count} appears {count} times between {start_num} and {end_num}.')
```

In this activity, we will create a custom function in Python to build a calculator that can perform basic arithmetic operations such as addition, subtraction, multiplication, and division.

```python
# Define a function called calculator that takes three arguments: num1, num2, and operation
def calculator(num1, num2, operation):
    if operation == '+':
        return num1 + num2
    elif operation == '-':
        return num1 - num2
    elif operation == '*':
        return num1 * num2
    elif operation == '/':

        # Handle division by zero case
        if num2 != 0:
            return num1 / num2
        else:
            return 'Error: Division by zero'
    else:
        return 'Invalid operation'
```

## **Bank Account Management System**

Create a Bank Account Management System using Object-Oriented Programming (OOPs) in Python. The system should allow users to create and manage bank accounts, deposit and withdraw funds, and view account details.

```python
# Define a BankAccount class
class BankAccount:
    def __init__(self, name, account_number, balance):
        self.name = name
        self.account_number = account_number
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        print(f'{amount} deposited. New balance is {self.balance}')

    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount
            print(f'{amount} withdrawn. New balance is {self.balance}')
        else:
            print(f'Insufficient balance. Current balance is {self.balance}')

    def account_details(self):
        print(f'Name: {self.name}\nAccount Number: {self.account_number}\nBalance: {self.balance}')

# Create a new instance of the BankAccount class with some initial balance
account = BankAccount(name='Kushagra Soni', account_number='9203****72', balance=566575350)

```

## **What's the time?**

Create a class Time which takes two inputs: hours and minutes to instantiate.

* Construct a method DisplayTime() which displays the time in AM/PM formats.

 * For example if the input is 14 hours and 45 mins, then this method will print "The time is 2:45 PM".
 * If the inputted hours exceeds 23 then print the message "The input hours should be less than 24" and if the inputted minutes exceeds 59 then print the message "The input minutes should be less than 60."
 * Also if the input is 12 hours 30 minutes, then the displayed time would be 12:30 PM

* Construct a method DisplayRatio() which should display the ratio of minutes to hours.
 * For example, (8 hours and 16 mins) should display 2. Use try, except block to account for ZeroDivisionError.

```python
class Time:
    def __init__(self, hours, minutes):
        self.hours = hours
        self.minutes = minutes

    def DisplayTime(self):
      try:
        if self.hours >= 24:
          print("Hours should not be exceed than 24 hours")
          return
        if self.minutes >=59:
          print("Minutes should not be exceed than 59 minutes")


          suffix = "AM"
          if self.hours >= 12:
              suffix = "PM"
              if self.hours > 12:
                  self.hours -= 12
          print(f"The time is {self.hours}:{self.minutes:02d} {suffix}")
      except Exception as e:
          print("Error:", e)

    def DisplayRatio(self):
        try:
            if self.hours == 0:
                raise ZeroDivisionError("Hours cannot be zero")
            ratio = self.minutes / self.hours
            print(f"The ratio of minutes to hours is {ratio:.2f}")
        except ZeroDivisionError as e:
            print("Error:", e)
```
